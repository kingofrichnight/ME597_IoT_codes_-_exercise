**<img src="https://github.com/purduelamm/purdue_me597_iiot/blob/main/lab/img/icon_SQL.png?raw=tru" width="40">SQL - SELECT statement for Practice #1**

```
SELECT
  timestamp AS "time",
  0+value as "Temperature"
FROM grafana_sample
WHERE
  $__timeFilter(timestamp) AND
  measurement = "temp" AND
  value != "UNAVAILABLE"
ORDER BY timestamp

```

1.	The reason why put ‘0+’ in front of ‘value’ column in the SELECT statement is to make the data in a numeric data type. We set up the ‘value’ column in the MySQL as ‘TINYTEXT’ that means basically return data is string data type.
2.	In addition, in the WHERE clause, ‘value != “UNAVAILABLE”’ makes your SELECT statement exclude “UNAVAILABE” data in the ‘value’ column.
3.	‘\$\__’ is indicating the macro function in Grafana to correlate the time range of Grafana interface with the database. ‘$__timeFilter(timestamp)’ means timestamps between the given time range. If you click ‘Show Help’ in the Query section, you will see the details of descriptions about the macro as below.

### Task 2.3

  1. Is it possible to build the IIoT monitoring system without MTConnect? Discuss the pros and cons of using MTConnect in the system.

Yes, it is absolutely possible to build an IIoT (Industrial Internet of Things) monitoring system without MTConnect. MTConnect is just one protocol/framework for standardizing data exchange between machines and software; you could instead use other protocols or custom solutions.

---
**Pros of MTConnect:**

1. Standardized format: XML-based data streams (current, sample, events) make it easy to parse and store.

2. Interoperability: Many machine tools and software platforms already support MTConnect → plug-and-play integration.

3. Extensibility: Can support new devices without major redesign.

4. Ease of monitoring: Tools like Grafana, MySQL, and Python scripts can easily consume MTConnect streams.

5. Open-source and widely adopted: Supported by research labs, universities, and industry.

---
**Cons of MTConnect:**

1. XML overhead: XML parsing can be slower and heavier than JSON or binary protocols for high-frequency data.

2. Limited adoption outside manufacturing: Not all industrial equipment supports MTConnect.

3. Learning curve: You need to understand MTConnect device streams, samples, and events.

4. Network dependency: MTConnect agents often require a stable network; adding many machines may require more infrastructure.

---
