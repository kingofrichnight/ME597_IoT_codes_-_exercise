### Q1. Please summarize Lab9.

---

Lab 9 focused on developing a complete real-time condition monitoring system for the air flow fixture (AFF) by integrating sensor data acquisition, machine learning, and industrial communication protocols. Acceleration data from the ADXL345 sensor was collected and processed to extract features, particularly using the X-axis RMS value to determine whether the AFF was running or stopped. An autoencoder-based anomaly detection model, developed in previous labs, was then used to classify the operating condition as normal or abnormal based on the MAE loss.

The system was extended by implementing an MTConnect adapter to stream real-time data, including RMS values, execution state, condition, and anomaly scores. The MTConnect agent was configured to receive this data and provide standardized XML outputs through endpoints such as /current. Finally, a data collector script was used to parse the MTConnect XML stream and store the data into a MySQL database for further analysis and visualization.

Overall, this lab demonstrated an end-to-end pipeline from sensor data acquisition to cloud-based storage, enabling real-time monitoring and diagnostics of machine conditions.
