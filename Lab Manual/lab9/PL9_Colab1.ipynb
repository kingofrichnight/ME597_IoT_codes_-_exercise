### Task 1.3 Using the list below, describe the data flow pipeline (system configuration) and software you need to use in order to realize real-time running condition prediction of the AFF.

  * Sensor: ADXL345
  * Middleware: MTConnect
  * Database: MySQL
  * Visualization: Grafana
  * Monitoring target(s):
     * Time domain features of each axis of the sensor
     * Anomaly prediction results of the AFF



---

The system starts from the ADXL345 accelerometer, which is attached to the AFF to measure vibration signals along the X, Y, and Z axes. The sensor is connected to a local processing device such as a microcontroller or Raspberry Pi, where the raw acceleration data is collected continuously in real time.

---
The acquired vibration data is then passed to a data acquisition and preprocessing program. In this stage, the software reads the sensor values, timestamps them, and calculates the required time-domain features for each axis. Typical features include mean, RMS, standard deviation, peak value, crest factor, and variance. These features represent the operating condition of the AFF more clearly than raw sensor samples alone.

---

After feature extraction, the processed data is sent through MTConnect middleware. MTConnect acts as the communication layer between the sensor-side processing program and higher-level monitoring software. It organizes the real-time data into a standardized structure so that the condition monitoring system can access the sensor features and machine status consistently.

---
Next, the feature data is fed into the prediction module, where a trained anomaly detection or classification model evaluates the current running condition of the AFF. The model outputs the anomaly prediction result, such as normal or abnormal condition, or a numeric anomaly score depending on the design.

---
Both the extracted time-domain features of each sensor axis and the anomaly prediction results are then stored in a MySQL database. MySQL works as the central storage system for real-time records, historical analysis, and dashboard queries. This allows the system to keep a full log of sensor behavior and prediction history.

---
Finally, Grafana is connected to the MySQL database for visualization. Grafana displays live dashboards showing the time-domain features of the X, Y, and Z axes as trends, gauges, or tables, along with the anomaly prediction results of the AFF. This enables users to monitor machine health in real time and quickly identify abnormal operating conditions.
