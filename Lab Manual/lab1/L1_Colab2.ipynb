### Task 2.1

Plot a chart using that includes both temperature unit in the code box below:

```python

df['time'] = pd.to_datetime(df["Timestamp"])
time = df['time'].dt.hour*3600 + df['time'].dt.minute*60 + df['time'].dt.second
t1 = time - time.iloc[0]

temperature1 = df['Temperature(°C)']
temperature2 = df['Temperature(°F)']

fig = plt.figure()
ax = fig.add_subplot(111)

ax.set_title('Temperature Graph', fontsize=20)
ax.set_xlabel('Seconds', fontsize=10)
ax.set_ylabel('Temperature', fontsize=10)

ax.plot(t1, temperature1, '-o', label='Celsius')
ax.plot(t1, temperature2, '-s', label='Fahrenheit')

ax.legend()

plt.show()
```
