## 2.1 Python datetime function

Python has a special date class, called “date.” A date, like a string, or a number, or a numpy array, has special rules for creating it and methods for working with it.

Here is an example on computing and printing local date and time:

````python
# Import datetime from the datetime module
from datetime import datetime

# Compute the local datetime: local_dt
local_dt = datetime.now()

# Print the local datetime
print(local_dt)
````

## 2.2 Formating date and time
You can also display the datetime function in different ways depending on convenience.
The following example illustrates how to personalize the date-time formatting:

````python
# Compute the local datetime: local_dt
local_dt = datetime.now().strftime("%Y-%m-%d %H:%M:%S.%f") #The string inside parenthesis define format for displs theaying. The % represents quantity, Y = year, m = month, d = day, etc.

# Print the local datetime
print(local_dt)
````

### Task 2.2
* Calculate how many days are left on this year.
* Determine if the processing time for printing longer strings in Python changes significantly in terms of seconds.

````python
from datetime import date

today = date.today()  # Get the current date
current_year = today.year # Get the current year
end_of_year = date(current_year, 12, 31) # Create a date object for the last day of the current year
days_remaining_timedelta = end_of_year - today # Calculate the difference between the end of the year and today
days_remaining = days_remaining_timedelta.days  # Extract the number of days from the timedelta object

print(f"There are {days_remaining} days remaining in the current year.")
````

````python
import time

starttime_short = time.time()
print("Hello")
endtime_short = time.time()
elapsed_time_short = endtime_short - starttime_short
print(f"Time taken for short string: {elapsed_time_short} seconds")

starttime_long = time.time()
print("Hello, I am AaryaFarheen from India. I study ms in autonomy in purdue main campus located in west laffayet")
endtime_long = time.time()
elapsed_time_long = endtime_long - starttime_long
print(f"Time taken for long string: {elapsed_time_long} seconds")

difference = ((elapsed_time_long) - (elapsed_time_short))
print(f"The difference in time is: {difference} seconds")
````
