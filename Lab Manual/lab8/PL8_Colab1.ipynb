## Foundation and Background

### What is Machine Learning (ML)?
Machine learning figures things out on its own. It doesn't need explicit programming to find patterns, features, or irregularities in big data sets. Compare the traditional programming paradigm to the machine learning paradigm below.

<img src="https://github.com/purduelamm/purdue_me597_iiot/blob/main/lab/img/prelab8_fig1.png?raw=true" width="100%">

*Figure 1 TP vs ML paradigm<sup>1</sup>*

For a Machine Learning algorithm to solve a problem, there are two key features. (1) it will take a piece of **data** and (2) will take a **guess** at the **answer**. Then, it will *optimize* its **guess** until it gets the correct **answer**. The concept of *loss* will be what helps to *optimize* the **guess**.





### Task 1.1
(1) Come up with an example of where Machine Learning would be more useful than Traditional Programming. (2) Explain why using Traditional Programming would not be optimal.



---

(1) Example where Machine Learning is more useful:
A good example is email spam detection. Instead of manually writing rules to classify emails as spam or not (e.g., checking for certain keywords like “free,” “win,” etc.), a Machine Learning model can be trained on thousands of labeled emails to automatically learn patterns. It can detect subtle features such as writing style, sender behavior, and hidden correlations that are difficult to explicitly program.

(2) Why Traditional Programming is not optimal:
Traditional programming relies on explicitly defined rules, which becomes inefficient in this case because spam emails constantly evolve. Writing rules to cover all variations would be extremely time-consuming and still incomplete. Additionally, rigid rules cannot adapt to new types of spam without continuous manual updates. In contrast, Machine Learning models can generalize from data and improve over time, making them more flexible, scalable, and accurate for such dynamic problems.

### Task 1.4

Looking at your visual graphs, what evidence do you have to support your belief whether the model is fit/underfit/overfit? Why does the model exhibit those fit-ness characteristics?



---

In the underfit model, the boundary is very simple (almost linear) and fails to correctly separate many data points from different classes. You can see several orange and purple points lying on the wrong side of the boundary. This indicates the model has not learned the underlying pattern of the data. The reason for this behavior is that the model is too simple (e.g., fewer layers, fewer neurons, or insufficient training), so it cannot capture the complexity of the dataset.

In the overfit model, the boundary is highly irregular and bends around individual data points. It closely follows even small fluctuations and noise in the dataset, often wrapping tightly around points. While it may classify training data almost perfectly, this complexity suggests poor generalization to new data. This happens because the model is too complex (e.g., too many neurons/layers or too many training epochs), causing it to memorize rather than learn general patterns.

In the fit model, the boundary is smooth and reasonably curved, separating the two classes with minimal misclassification. It captures the overall structure of the data without reacting to noise. This indicates a good balance between model complexity and training, allowing it to generalize well.
