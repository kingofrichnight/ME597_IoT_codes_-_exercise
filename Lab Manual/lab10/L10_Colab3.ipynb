## Possible Ways to Fix Sporadic False Predictions

- Use **prediction smoothing / majority voting** over several consecutive outputs instead of using only one instant prediction. This can reduce sudden label changes caused by noise.  

- Improve the **training dataset quality and balance** by collecting more sound samples for each class, especially for confusing states such as *Air leaking* and *Vacuuming*. More diverse data can help the model generalize better.  

- Refine the **feature extraction process**, such as testing different spectrogram settings, window sizes, overlap, or additional audio features, so that the difference between classes becomes clearer to the model.  

- Try a more suitable **deep learning architecture** or tune hyperparameters. For example, adjusting the number of CNN layers, filter sizes, dropout, or using CNN-LSTM models may improve temporal understanding of sound patterns.  

- Add a **confidence threshold or decision logic** so that predictions only change when the model is sufficiently confident for several consecutive time steps. This helps avoid rapid switching between classes.
